DATA TYPE CORRECTION

```python
data['gender'] = pd.to_numeric(data['gender'].str.replace(',', ''), errors='coerce')
data.head(2)
```

STANDARDIZATION

```python
data['loyalty_status'] = data['loyalty_status'].str.lower().str.strip()
data.head(2)
```

```python
data['region'] = data['region'].replace({
    'north': 'NORTH',
    'North': 'NORTH',
    'south': 'SOUTH',
    'South': 'SOUTH'
})
```
